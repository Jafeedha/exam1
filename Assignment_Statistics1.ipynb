4. What do you think about the scoring variation for math, reading and writing
based on

â— Gender

```python
df[['math score','reading score','writing score']].agg([np.mean,np.std,np.max,np.min])
```

```python
df.groupby(['gender'])[['math score','reading score','writing score']].agg([np.mean,np.std,np.max,np.min])
```

   In every subject  there is students with 100% marks in both male and female

â— Test preparation course

```python
df.groupby(['test preparation course'])[['math score','reading score','writing score']].agg([np.mean,np.std,np.max,np.min])
```
